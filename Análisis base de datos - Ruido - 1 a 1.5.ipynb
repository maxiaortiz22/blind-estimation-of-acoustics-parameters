```python
#Me muevo al directorio de las bases de datos:
os.chdir('/content/drive/My Drive/Tesis/Base de datos/TAE sinteticos y reales - 1 a 1.5s')

with open(f'Parámetros/Parametros_125.pickle', 'rb') as handle:
    params_125 = pickle.load(handle)
with open(f'Parámetros/Parametros_250.pickle', 'rb') as handle:
    params_250 = pickle.load(handle)
with open(f'Parámetros/Parametros_500.pickle', 'rb') as handle:
    params_500 = pickle.load(handle)
with open(f'Parámetros/Parametros_1000.pickle', 'rb') as handle:
    params_1000 = pickle.load(handle)
with open(f'Parámetros/Parametros_2000.pickle', 'rb') as handle:
    params_2000 = pickle.load(handle)
with open(f'Parámetros/Parametros_4000.pickle', 'rb') as handle:
    params_4000 = pickle.load(handle)
with open(f'Parámetros/Parametros_8000.pickle', 'rb') as handle:
    params_8000 = pickle.load(handle)

names = list(params_125.keys())
```

```python
# T30:
t30_125, t30_250, t30_500 = [], [], []
t30_1000, t30_2000, t30_4000, t30_8000 = [], [], [], []

for name in names:
  t30_125.append(params_125[name][0])
  t30_250.append(params_250[name][0])
  t30_500.append(params_500[name][0])
  t30_1000.append(params_1000[name][0])
  t30_2000.append(params_2000[name][0])
  t30_4000.append(params_4000[name][0])
  t30_8000.append(params_8000[name][0])
```

```python
# C50:
c50_125, c50_250, c50_500 = [], [], []
c50_1000, c50_2000, c50_4000, c50_8000 = [], [], [], []

for name in names:
  c50_125.append(params_125[name][1])
  c50_250.append(params_250[name][1])
  c50_500.append(params_500[name][1])
  c50_1000.append(params_1000[name][1])
  c50_2000.append(params_2000[name][1])
  c50_4000.append(params_4000[name][1])
  c50_8000.append(params_8000[name][1])
```

```python
# C80:
c80_125, c80_250, c80_500 = [], [], []
c80_1000, c80_2000, c80_4000, c80_8000 = [], [], [], []

for name in names:
  c80_125.append(params_125[name][2])
  c80_250.append(params_250[name][2])
  c80_500.append(params_500[name][2])
  c80_1000.append(params_1000[name][2])
  c80_2000.append(params_2000[name][2])
  c80_4000.append(params_4000[name][2])
  c80_8000.append(params_8000[name][2])
```

```python
# D50:
d50_125, d50_250, d50_500 = [], [], []
d50_1000, d50_2000, d50_4000, d50_8000 = [], [], [], []

for name in names:
  d50_125.append(params_125[name][3])
  d50_250.append(params_250[name][3])
  d50_500.append(params_500[name][3])
  d50_1000.append(params_1000[name][3])
  d50_2000.append(params_2000[name][3])
  d50_4000.append(params_4000[name][3])
  d50_8000.append(params_8000[name][3])
```

# Elimino audios con Tr que no esté entre 1 y 1.5 segundos

Saco los audios que no tengan un Tr entre 1 y 1.5 segundos, usando la banda de 1 kHz como referencia

```python
with open(f'Parámetros/Parametros_1000.pickle', 'rb') as handle:
    params_1000 = pickle.load(handle)

names = list(params_1000.keys())

t30_1000 = []
for name in names:
  t30_1000.append(params_1000[name][0])
print(t30_1000[0])
```
